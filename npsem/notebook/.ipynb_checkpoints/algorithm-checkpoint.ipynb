```python
from itertools import product

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tqdm
from joblib import Parallel, delayed

from npsem.model import CD, SCM
from npsem.model_utils import random_scm

import npsem.pomis_plus as pomp
```

#### Step 1
**1. Define a static graph for each time slice**

**2. Define a transition edges**

```python
# static graph
G = CD(['X', 'Z', 'Y'],
       [('X', 'Z'), ('Z', 'Y')],
       [('X', 'Y', 'U_XY')])

# transition edges : can include DUC too
transition_edges = [('X0', 'X1'), ('X1', 'X2')]
```

```python
# multi time step graph from static graph
def create_multiple_timestep_graph(G: CD, transition_edges: list[tuple], t: int):
    variables = G.V
    UCs = G.confounded_to_3tuples()
    edges = G.edges
    
    print(f"static UCs from confounded_to_3tuples() : {UCs}")
    new_variables = []
    new_UCs = []
    new_edges = []
    
    for i in range(t):
        for v in variables:
            new_variables.append(f"{v}{i}")
        for var_a, var_b, UC in UCs:
            new_UCs.append((f"{var_a}{i}", f"{var_b}{i}", f"{UC}_{i}"))
        for edge in edges:
            new_edges.append((f"{edge[0]}{i}", f"{edge[1]}{i}"))
    
    new_G = CD(new_variables, new_edges, new_UCs)
    print(new_G)
    new_G += transition_edges 

    return new_G
```

```python
# create multiple timestep graph
multiple_G = create_multiple_timestep_graph(G, transition_edges, 3)
print(multiple_G)

# random SCM for convenience
multiple_M, _ = random_scm(multiple_G, seed=1)
print(multiple_M)
print(_)
# reward variables
reward_vars = ['Y0', 'Y1', 'Y2']
```

#### Step 2
**1. Find set of arms (POMIS hopefully)**
- For this example, we assume that the tree below is correct a set of arms that include optimal arms  

**2. Calculate expected rewards of each arm -> finding optimal arm to calculate regret**
  
**3. Generate all data in advance**
- In our setting, it seems like observing each reward variable along each intervention step by step is equivalent to observing all reward variables after intervening all action variables simultaneously. Is this correct? 
- If it is true, we can sample all data of each actions in advance for convenience
- If it is not correct, I think it is not allowed to create data in advance as we did, so we need to change that part

```python
# calculate action sequences(POMIS+) from the POMIS+ algorithm

# set of arms according to the tree above
# the number of actions = the number of leaves of the tree / # of 72
arms = [('Z0', 'Z1', 'Z2'),
        ('Z0', 'Z1', 'X2', 'Z2'),
        ('Z0', 'X1', 'Z1'),
        ('Z1', 'Z2'),
        ('Z1', 'X2', 'Z2'),
        ('X1', 'Z1'),
        ('X0', 'Z0', 'Z2'),
        ('X0', 'Z0', 'X2', 'Z2')]
```

```python
def create_all_actions(arms: list[tuple]) -> list[dict] :
    """
    create all binary actions given the set of arms
    """
    arms.sort()
    all_actions = []
    for arm in arms:
        # [(0,1), (1,0), (1,1), (0,0)] if len(arm) == 2
        for comb in product((0, 1), repeat=len(arm)):
            action = dict(zip(arm, comb)) # e.g., {"Z0": 0, "Z1": 1}...
            all_actions.append(action)
    return all_actions # [{"Z0": 0, "Z1": 1}, {"Z0": 1, "Z1": 1}, ... ]
    
actions = create_all_actions(arms)
print(len(actions))
for idx, action in enumerate( actions ):
    print(f"{idx+1}: { action }")
```

```python
# calculate true expected rewards from above action sequences.
# E[Y0+Y1+Y2]

# To calculate regret, we need to find the optimal arm among all possible arms under current SCM 
# For now, we consider if the optimal arm is in the tree
def calculate_expected_rewards(M: SCM, actions: list[dict], reward_vars: list[str]) -> list[float]:
    """
    calculate the expected reward for each action       
    """
    # ['Y0', 'Y1', 'Y2']
    reward_vars = tuple(reward_vars)
    # [0,0,0,...,] (0 with number of action sequences)
    expected_rewards = np.zeros(len(actions))

    #one of those action seq. (e.g., [..., {'X0': 0, 'Z0': 0, 'X2': 1, 'Z2': 0}, ...]) 
    for idx, action in enumerate(actions):
        # get P(reward_vars | do (action)) 
        result = M.query(outcome=reward_vars, intervention=action)
        # if idx == 0 : print(result)
        val = 0
        for k, v in result.items(): # k: values of Ys, v: prob. of the values
            val += sum(k)*v
        expected_rewards[idx] = val
    return expected_rewards 

print(calculate_expected_rewards(multiple_M, actions, reward_vars))
```

```python
# For saving times, we create all data corresponding to each action in advance
def sampling_from_SCM(M: SCM, size: int, intervention: dict) -> dict:
    """
    sample from SCM given the intervention
    """
    U = list(sorted(M.G.U | M.more_U))  # all exogenous variables
    D = M.D # domain of each variable       
    P_U = M.P_U # distribution of exogenous variables
    V_ordered = M.G.causal_order()

    u_vals = {tuple(u): P_U(dict(zip(U, u))) for u in product(*[D[U_i] for U_i in U])}
    u_configs = tuple(u_vals.keys())  # order preserved! 3.7!
    unit_counts = np.random.multinomial(size, list(u_vals.values()))
    data = np.zeros((size, len(V_ordered)))
    offset = 0
    for ith_unit, how_many in enumerate(unit_counts):
        if how_many == 0:
            continue
        u = u_configs[ith_unit]
        assigned = dict(zip(U, u))
        for V_i in V_ordered:
            if V_i in intervention:
                assigned[V_i] = intervention[V_i]
            else:
                assigned[V_i] = M.F[V_i](assigned)

        generated = np.array([assigned[V_i] for V_i in V_ordered])
        data[offset:(offset + how_many), :] = generated
        offset += how_many
    np.random.shuffle(data)
    sampled_data = pd.DataFrame(data, columns=V_ordered)
    return sampled_data

print(sampling_from_SCM(multiple_M, 10, {'X0': 1, 'X1': 0, 'Z2': 1}))

def data_generator(M: SCM, actions: list[dict], T: int) -> dict[int, pd.DataFrame]:
    """
    generate data from SCM given the actions
    """
    all_data = dict()
    for idx, action in enumerate( actions ):
        sampled_data = sampling_from_SCM(M, T, action)
        all_data[idx] = sampled_data
    
    return all_data
```

#### Step 3
**1. UCB algorithm**

**2. Run simulation** 

```python
def UCB(data: dict, actions: list[dict], reward_vars: list[str]) -> np.ndarray:
    # the number of playing of one policy
    T = len(data[0])
    
    # means, counts for each action sequence [mean for action 0, mean for action 1, ...]
    actions_means = np.zeros(len(actions))
    actions_counts = np.zeros(len(actions), dtype=int)
    action_played = np.zeros(T, dtype=int)
    
    for t in range(T):
        ucb_values = actions_means + np.sqrt(2 * np.log(t + 1) / np.maximum(actions_counts, 1))
        ucb_values[actions_counts == 0] = np.inf  
        action_index = np.argmax(ucb_values)  
        
        # data[action_index] : insert key into dict 'data' & ['Y1', 'Y2', 'Y3'] from line t
        reward = data[action_index][reward_vars].iloc[t, :] 
            
        # sum of all rewards
        total_reward = sum(reward)
        
        # +1 count for the action
        actions_counts[action_index] += 1
        
        actions_means[action_index] = actions_means[action_index] + (total_reward - actions_means[action_index]) / actions_counts[action_index]
        
        action_played[t] = action_index

    return action_played
```

```python
def kl_ucb(data: dict, actions: list[dict], reward_vars: list[str], c: float = 1.0) -> np.ndarray:
    """
    KL-UCB alg.
    Parameters:
        data : 각 action의 intervened data
        actions : 액션 시퀀스 리스트
        reward_vars : 보상 변수들
        c : 탐험의 정도 parameter
    Returns:
        action_played (np.ndarray): 각 시간 t에서 선택된 액션 인덱스
    """
    # 총 시간 단계
    T = len(data[0])

    # 각 액션의 평균 보상과 선택 횟수 초기화
    actions_means = np.zeros(len(actions))
    actions_counts = np.zeros(len(actions), dtype=int)
    action_played = np.zeros(T, dtype=int)
    
    def kl_divergence(p, q):
        # KL Divergence 계산 (이항 분포)
        epsilon = 1e-15  # 안정성을 위한 작은 값
        p = np.clip(p, epsilon, 1 - epsilon)
        q = np.clip(q, epsilon, 1 - epsilon)
        return p * np.log(p / q) + (1 - p) * np.log((1 - p) / (1 - q))
    
    def find_upper_bound(mean, log_term, tol=1e-4):
        #KL-UCB 점수 계산을 위한 q 찾기 (이항 분포 상한값)
        lower, upper = mean, 1.0  # q는 평균 ~ 1 사이에 존재..
        while upper - lower > tol:
            mid = (lower + upper) / 2
            if kl_divergence(mean, mid) > log_term:
                upper = mid
            else:
                lower = mid
        return (lower + upper) / 2
    
    for t in range(T):
        ucb_values = np.zeros(len(actions))
        log_term = np.log(t + 1) + c * np.log(np.log(t + 2))
        
        # 각 액션의 KL-UCB 점수 계산
        for i in range(len(actions)):
            if actions_counts[i] == 0:
                ucb_values[i] = np.inf
            else:
                ucb_values[i] = find_upper_bound(actions_means[i], log_term / actions_counts[i])
        
        # UCB 값이 최대인 액션 선택
        action_index = np.argmax(ucb_values)
        
        # 데이터에서 보상 가져오기
        reward = data[action_index][reward_vars].iloc[t, :]
        total_reward = sum(reward)
        
        # 선택 횟수와 평균 보상 업데이트
        actions_counts[action_index] += 1
        actions_means[action_index] += (total_reward - actions_means[action_index]) / actions_counts[action_index]
        
        action_played[t] = action_index
    
    return action_played
```

```python
def thompson_sampling(data: dict, actions: list[dict], reward_vars: list[str]) -> np.ndarray:
    """
    Thompson Sampling 알고리즘
    Parameters:
        data : 액션별 보상 데이터
        actions : 액션 시퀀스 리스트
        reward_vars : 보상 변수들
    Returns:
        action_played (np.ndarray): 각 시간 t에서 선택된 액션 인덱스
    """
    # 총 시간 단계
    T = len(data[0])

    # Beta 분포의 파라미터 초기화 (각 arm에 대해 성공 횟수 alpha와 실패 횟수 beta)
    alpha = np.ones(len(actions))  # 성공 횟수 (초기값 1)
    beta = np.ones(len(actions))   # 실패 횟수 (초기값 1)

    action_played = np.zeros(T, dtype=int)

    for t in range(T):
        # 각 arm에 대해 Beta 분포에서 샘플링
        sampled_theta = np.random.beta(alpha, beta)

        # 샘플링된 theta 값이 가장 큰 arm 선택
        action_index = np.argmax(sampled_theta)

        # 데이터에서 보상 가져오기
        reward = data[action_index][reward_vars].iloc[t, :]
        total_reward = sum(reward)

        # Beta 분포의 파라미터 업데이트
        if total_reward > 0:  # 성공 시 alpha 증가
            alpha[action_index] += 1
        else:  # 실패 시 beta 증가
            beta[action_index] += 1

        action_played[t] = action_index

    return action_played
```

```python
# example for generating data and running the UCB algorithm with the data
data = data_generator(multiple_M, actions, 100)
print(data) # {'0': (100x9) table, ...} '0' is one action sequence
print(UCB(data, actions, reward_vars))
print(kl_ucb(data, actions, reward_vars))
print(thompson_sampling(data, actions, reward_vars))
```

```python
# the played arm 
def calculate_expected_regret(expected_rewards, arm_played):
    """
    calculate the expected regret 
    """
    T = len(arm_played)
    reward = 0
    for t in range(T):
        # expected_rewards of each arm = [0.60826,  1.00346 , ...] from SCM
        # 해당 t번째 action (arm_played[t]), arm_played= [ 0  1  2  3  4 ...] from UCB alg.
        
        # sum of `expected rewards' of all played arm
        reward += expected_rewards[arm_played[t]]
        
    # find the `best reward' among all possible arms
    best_reward = np.max(expected_rewards)
        
    # calculate the cumulative regret
    cumulative_regret = T * best_reward - reward

    return cumulative_regret
```

```python
def single_simulation_ucb1(Tvals, M, actions, reward_vars):
    # Tvals : the number of data generated
    ucb_expected_regrets = np.zeros(len(Tvals))
    
    for i, T in tqdm.tqdm(enumerate(Tvals), total=len(Tvals), desc="Single Simulation"): 
        data = data_generator(M, actions, T)
        expected_rewards = calculate_expected_rewards(M, actions, reward_vars)
        
        # ucb1 regert
        ucb_arms = UCB(data, actions, reward_vars)
        ucb_expected_regrets[i] = calculate_expected_regret(expected_rewards, ucb_arms)

    return ucb_expected_regrets
```

```python
def single_simulation_klucb(Tvals, M, actions, reward_vars):
    # Tvals : the number of data generated
    kl_ucb_expected_regrets = np.zeros(len(Tvals))
    
    for i, T in tqdm.tqdm(enumerate(Tvals), total=len(Tvals), desc="Single Simulation"): 
        data = data_generator(M, actions, T)
        expected_rewards = calculate_expected_rewards(M, actions, reward_vars)
        
        # kl-ucb regert
        ucb_arms = kl_ucb(data, actions, reward_vars)
        kl_ucb_expected_regrets[i] = calculate_expected_regret(expected_rewards, ucb_arms)

    return kl_ucb_expected_regrets
```

```python
def single_simulation_TS(Tvals, M, actions, reward_vars):
    # Tvals : the number of data generated
    TS_expected_regrets = np.zeros(len(Tvals))
    
    for i, T in tqdm.tqdm(enumerate(Tvals), total=len(Tvals), desc="Single Simulation"): 
        data = data_generator(M, actions, T)
        expected_rewards = calculate_expected_rewards(M, actions, reward_vars)
        
        # kl-ucb regert
        ucb_arms = thompson_sampling(data, actions, reward_vars)
        TS_expected_regrets[i] = calculate_expected_regret(expected_rewards, ucb_arms)

    return TS_expected_regrets
```

```python
def run_simulation(Tvals: np.ndarray, M: SCM, actions: list[dict], reward_vars: list[str], avg: int):
    # Parallel processing
    results = Parallel(n_jobs=-1)(
        delayed(single_simulation_klucb)(Tvals, M, actions, reward_vars) for _ in tqdm.tqdm(range(avg), desc="Running Simulations")
    )
    
    print(f"All simulation complete : {results}")
    
    solver_expected_regrets = np.sum(results, axis=0)
    avg_solver_expected_regrets = solver_expected_regrets / avg

    return avg_solver_expected_regrets

def plot_regrets(Tvals, results):
    plt.figure()
    plt.plot(Tvals, results, label="UCB", color="blue", marker="s", markersize=3, linestyle="-", linewidth=1)
    plt.title("Cumulative regret")
    plt.xlabel("Trials")
    plt.ylabel("Cum. Regrets")
    plt.legend(loc="upper left")
    plt.grid(True)
    plt.savefig("../figures/algo_results/UCB.png")
    plt.show()
```

```python
# Simulation
Tvals = np.arange(1000, 15000, 1000) # 15000
avg = 10 #100
print(Tvals)
```

```python
results = run_simulation(Tvals, multiple_M, actions, reward_vars, avg)
plot_regrets(Tvals, results)
```
